# Single Qubit Gates with Q5M.js

In this notebook, we'll explore the fundamental single-qubit gates in quantum computing. These gates are the building blocks for more complex quantum operations.

## The Pauli Gates

The Pauli gates (X, Y, Z) are fundamental single-qubit gates:

- **X Gate (Bit Flip)**: $X|0\rangle = |1\rangle$ and $X|1\rangle = |0\rangle$
- **Y Gate**: $Y|0\rangle = i|1\rangle$ and $Y|1\rangle = -i|0\rangle$
- **Z Gate (Phase Flip)**: $Z|0\rangle = |0\rangle$ and $Z|1\rangle = -|1\rangle$

Let's start by exploring the X gate:

```javascript
import { Circuit } from 'q5m';

console.log('Q5M.js library loaded successfully!');

// Create a circuit and apply X gate
const circuit = new Circuit(1); // Create circuit with 1 qubit starting with |0⟩

console.log('\nPauli X Gate (Bit Flip):');
let state = circuit.execute();
console.log(`Initial state: |0⟩: 100.0% probability`);
console.log('State representation:', state.toString());

// Apply X gate
circuit.x(0); // lowercase x
state = circuit.execute();
console.log(`After X gate:  |1⟩: 100.0% probability`);
console.log('State representation:', state.toString());

console.log('\nThe X gate flipped |0⟩ to |1⟩!');
```

Perfect! The X gate successfully flipped our qubit from $|0\rangle$ to $|1\rangle$.

## The Y Gate

The Y gate combines a bit flip with a phase change. It introduces complex amplitudes:

```javascript
// Reset circuit for Y gate demonstration
const yCircuit = new Circuit(1); // Create circuit with 1 qubit

console.log('Pauli Y Gate:');
let yState = yCircuit.execute();
console.log(`Starting from |0⟩: 100.0% probability`);
console.log('State representation:', yState.toString());

// Apply Y gate
yCircuit.y(0); // lowercase y
yState = yCircuit.execute();
console.log(`After Y gate: |1⟩: 100.0% probability with imaginary amplitude`);
console.log('State representation:', yState.toString());

console.log('\nNotice the complex amplitudes! Y|0⟩ = i|1⟩');
```

## The Z Gate (Phase Flip)

The Z gate leaves $|0\rangle$ unchanged but adds a phase of -1 to $|1\rangle$. Let's see this in action:

```javascript
console.log('Pauli Z Gate (Phase Flip):');

// Test Z gate on |0⟩
const z1Circuit = new Circuit(1); // |0⟩ state

console.log('Z gate on |0⟩:');
let z1State = z1Circuit.execute();
console.log(`Before: |0⟩ state`);
console.log('State representation:', z1State.toString());

z1Circuit.z(0); // lowercase z
z1State = z1Circuit.execute();
console.log(`After:  |0⟩ state (unchanged)`);
console.log('State representation:', z1State.toString());
console.log('No change for |0⟩!');

// Test Z gate on |1⟩
const z2Circuit = new Circuit(1);
z2Circuit.x(0); // Flip to |1⟩ first

console.log('\nZ gate on |1⟩:');
let z2State = z2Circuit.execute();
console.log(`Before: |1⟩ state`);
console.log('State representation:', z2State.toString());

z2Circuit.z(0); // lowercase z
z2State = z2Circuit.execute();
console.log(`After:  |1⟩ state with -1 phase`);
console.log('State representation:', z2State.toString());
console.log('Z gate added a -1 phase to |1⟩!');
```

## The Hadamard Gate Revisited

We've seen the Hadamard gate before. It creates superposition states:

$$H = \frac{1}{\sqrt{2}}\begin{pmatrix}1 & 1\\1 & -1\end{pmatrix}$$

```javascript
console.log('Hadamard Gate:');

// H|0⟩ = |+⟩
const h1Circuit = new Circuit(1);
h1Circuit.h(0); // lowercase h

console.log('H|0⟩ creates |+⟩ = (|0⟩ + |1⟩)/√2');
let h1State = h1Circuit.execute();
console.log('After H: Equal superposition state');
console.log('State representation:', h1State.toString());
console.log(`Probabilities: 50.0% each`);

// H|1⟩ = |-⟩
const h2Circuit = new Circuit(1);
h2Circuit.x(0); // Start with |1⟩
h2Circuit.h(0); // lowercase h

console.log('\nH|1⟩ creates |-⟩ = (|0⟩ - |1⟩)/√2');
let h2State = h2Circuit.execute();
console.log('After H: Equal superposition with opposite phases');
console.log('State representation:', h2State.toString());
console.log(`Probabilities: 50.0% each, but with opposite phases!`);
```

## Phase Gates (S and T)

The S and T gates are important phase gates:

- **S Gate**: Adds a phase of $i$ to $|1\rangle$
- **T Gate**: Adds a phase of $e^{i\pi/4}$ to $|1\rangle$

```javascript
console.log('Phase Gates:');

// S Gate
const sCircuit = new Circuit(1);
sCircuit.x(0); // Start with |1⟩

console.log('S Gate (π/2 phase):');
let sState = sCircuit.execute();
console.log(`Before S: |1⟩ state`);
console.log('State representation:', sState.toString());

sCircuit.s(0); // lowercase s
sState = sCircuit.execute();
console.log(`After S:  |1⟩ state with i phase`);
console.log('State representation:', sState.toString());
console.log('S gate added i phase to |1⟩!');

// T Gate
const tCircuit = new Circuit(1);
tCircuit.x(0); // Start with |1⟩

console.log('\nT Gate (π/4 phase):');
let tState = tCircuit.execute();
console.log(`Before T: |1⟩ state`);
console.log('State representation:', tState.toString());

tCircuit.t(0); // lowercase t
tState = tCircuit.execute();
console.log(`After T:  |1⟩ state with e^(iπ/4) phase`);
console.log('State representation:', tState.toString());
console.log('T gate added e^(iπ/4) phase to |1⟩!');
```

## Rotation Gates

Q5M.js also provides rotation gates around the X, Y, and Z axes:

- **Rx(θ)**: Rotation around X-axis
- **Ry(θ)**: Rotation around Y-axis  
- **Rz(θ)**: Rotation around Z-axis

```javascript
console.log('Rotation Gates:');

// Rx rotation
const rxCircuit = new Circuit(1);
console.log('Rx(π/2) - 90° rotation around X-axis:');
let rxState = rxCircuit.execute();
console.log(`Before: |0⟩ state`);
console.log('State representation:', rxState.toString());

rxCircuit.rx(0, Math.PI/2); // lowercase rx
rxState = rxCircuit.execute();
console.log(`After:  Mixed state with complex amplitudes`);
console.log('State representation:', rxState.toString());

// Ry rotation  
const ryCircuit = new Circuit(1);
console.log('\nRy(π/4) - 45° rotation around Y-axis:');
let ryState = ryCircuit.execute();
console.log(`Before: |0⟩ state`);
console.log('State representation:', ryState.toString());

ryCircuit.ry(0, Math.PI/4); // lowercase ry
ryState = ryCircuit.execute();
console.log(`After:  Mixed state (mostly |0⟩ with some |1⟩)`);
console.log('State representation:', ryState.toString());

// Rz rotation
const rzCircuit = new Circuit(1);
rzCircuit.x(0); // Start with |1⟩
console.log('\nRz(π/3) - 60° rotation around Z-axis:');
let rzState = rzCircuit.execute();
console.log(`Before: |1⟩ state`);
console.log('State representation:', rzState.toString());

rzCircuit.rz(0, Math.PI/3); // lowercase rz
rzState = rzCircuit.execute();
console.log(`After:  |1⟩ state with phase rotation`);
console.log('State representation:', rzState.toString());
```

## Summary

In this notebook, we explored the fundamental single-qubit gates:

1. **Pauli Gates**:
   - X: Bit flip ($|0\rangle \leftrightarrow |1\rangle$)
   - Y: Bit flip with phase ($|0\rangle \to i|1\rangle$, $|1\rangle \to -i|0\rangle$)
   - Z: Phase flip ($|1\rangle \to -|1\rangle$)

2. **Hadamard Gate**: Creates superposition states

3. **Phase Gates**:
   - S: Adds $i$ phase to $|1\rangle$
   - T: Adds $e^{i\pi/4}$ phase to $|1\rangle$

4. **Rotation Gates**: Rx, Ry, Rz for arbitrary rotations

These gates form the foundation for building more complex quantum circuits and algorithms!
