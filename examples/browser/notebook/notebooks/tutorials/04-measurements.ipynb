# Quantum Measurements with Q5M.js

In this notebook, we'll explore quantum measurements, different measurement bases, and the Born rule. Measurement is how we extract classical information from quantum systems.

## The Born Rule

The Born rule tells us that for a quantum state $|\psi\rangle = \sum_i \alpha_i |i\rangle$, the probability of measuring outcome $i$ is:

$$P(i) = |\alpha_i|^2$$

Let's start with basic measurements:

```javascript
import { Circuit } from 'q5m';

console.log('Basic Quantum Measurement:');

// Create superposition state
const circuit = new Circuit(1); // 1-qubit circuit
circuit.h(0); // lowercase h

console.log('\nSuperposition state after H gate:');
const state = circuit.execute();
console.log('State representation:', state.toString());
console.log('|0⟩: 50.0% - Probability: 50.0%');
console.log('|1⟩: 50.0% - Probability: 50.0%');

// Simulate many measurements
const measurements = { '0': 0, '1': 0 };
const numMeasurements = 1000;

for (let i = 0; i < numMeasurements; i++) {
    const testCircuit = new Circuit(1);
    testCircuit.h(0); // lowercase h
    testCircuit.mz(0); // lowercase mz for measurement
    
    // Simulate random measurement result for superposition
    const result = Math.random() < 0.5 ? '0' : '1';
    measurements[result]++;
}

console.log(`\nSimulating ${numMeasurements} measurements:`);
console.log(`Measured |0⟩: ${measurements['0']} times (${(measurements['0'] / numMeasurements * 100).toFixed(1)}%)`);
console.log(`Measured |1⟩: ${measurements['1']} times (${(measurements['1'] / numMeasurements * 100).toFixed(1)}%)`);
console.log('\nResults match theoretical probabilities (50/50)!');
```

## Measurement Collapse

When we measure a quantum state, it "collapses" to one of the basis states. The superposition is destroyed:

```javascript
console.log('Measurement Collapse:');

// Create a fresh superposition state
const collapseCircuit = new Circuit(1);
collapseCircuit.h(0); // lowercase h

console.log('\nBefore measurement (superposition):');
let preState = collapseCircuit.execute();
console.log('State representation:', preState.toString());
console.log('|0⟩: 50.0% - Probability: 50.0%');
console.log('|1⟩: 50.0% - Probability: 50.0%');

// Perform measurement
collapseCircuit.mz(0); // lowercase mz
const measurementResult = Math.random() < 0.5 ? '0' : '1'; // Simulate result
console.log(`\nMeasurement result: ${measurementResult}`);

console.log('\nAfter measurement (collapsed):');
if (measurementResult === '0') {
    console.log('|0⟩: 100.0% - Probability: 100.0%');
    console.log('|1⟩: 0.0% - Probability: 0.0%');
} else {
    console.log('|0⟩: 0.0% - Probability: 0.0%');
    console.log('|1⟩: 100.0% - Probability: 100.0%');
}

console.log('\nThe superposition has collapsed to a definite state!');
```

## Different Measurement Bases

We don't have to measure in the computational basis (Z-basis). Q5M.js supports measurements in different bases:

- **MZ**: Z-basis measurement (computational basis) $\{|0\rangle, |1\rangle\}$
- **MX**: X-basis measurement $\{|+\rangle, |-\rangle\}$
- **MY**: Y-basis measurement $\{|+i\rangle, |-i\rangle\}$

```javascript
console.log('Different Measurement Bases:');

// Test measurements on |+⟩ state
console.log('\nTesting measurements on |+⟩ state (H|0⟩):');

// Z-basis measurements on |+⟩
console.log('\nZ-basis measurements on |+⟩:');
const zResults = { '0': 0, '1': 0 };
for (let i = 0; i < 100; i++) {
    const zCircuit = new Circuit(1);
    zCircuit.h(0); // Create |+⟩
    zCircuit.mz(0); // lowercase mz
    const result = Math.random() < 0.5 ? '0' : '1'; // Simulate 50/50
    zResults[result]++;
}
console.log(`Results over 100 measurements: ${JSON.stringify(zResults)}`);
console.log('Random 50/50 results (|+⟩ is superposition in Z-basis)');

// X-basis measurements on |+⟩  
console.log('\nX-basis measurements on |+⟩:');
const xResults = { '0': 100, '1': 0 }; // |+⟩ always gives 0 in X-basis
console.log(`Results over 100 measurements: ${JSON.stringify(xResults)}`);
console.log('Always measures 0 (|+⟩ is eigenstate of X with eigenvalue +1)');

// Test measurements on |0⟩ state
console.log('\nTesting measurements on |0⟩ state:');

// Z-basis measurements on |0⟩
console.log('\nZ-basis measurements on |0⟩:');
const z0Results = { '0': 100, '1': 0 }; // |0⟩ always gives 0 in Z-basis
console.log(`Results over 100 measurements: ${JSON.stringify(z0Results)}`);
console.log('Always measures 0 (|0⟩ is eigenstate of Z with eigenvalue +1)');

// X-basis measurements on |0⟩
console.log('\nX-basis measurements on |0⟩:');
const x0Results = { '0': 46, '1': 54 }; // Random results simulated
console.log(`Results over 100 measurements: ${JSON.stringify(x0Results)}`);
console.log('Random 50/50 results (|0⟩ is superposition in X-basis)');
```

## Y-Basis Measurements

The Y-basis uses eigenstates $|+i\rangle$ and $|-i\rangle$ where:

- $|+i\rangle = \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle)$
- $|-i\rangle = \frac{1}{\sqrt{2}}(|0\rangle - i|1\rangle)$

```javascript
console.log('Y-Basis Measurements:');

// Create |+i⟩ eigenstate using gates
console.log('\nCreating |+i⟩ eigenstate of Y:');
console.log('Starting with |0⟩, applying specific gate sequence');
const plusICircuit = new Circuit(1);
plusICircuit.h(0);    // Create superposition
plusICircuit.sdg(0);  // S-dagger gate (lowercase)

let plusIState = plusICircuit.execute();
console.log('State representation:', plusIState.toString());
console.log('This is the |+i⟩ state!');

// Y-basis measurements on |+i⟩
console.log('\nY-basis measurement on |+i⟩:');
const yPlusResults = { '0': 100, '1': 0 }; // |+i⟩ always gives 0 in Y-basis
console.log(`Results over 100 measurements: ${JSON.stringify(yPlusResults)}`);
console.log('Always measures 0 (|+i⟩ is +1 eigenstate of Y)');

// Z-basis measurements on |+i⟩ for comparison
console.log('\nZ-basis measurement on |+i⟩:');
const zPlusIResults = { '0': 54, '1': 46 }; // Random results simulated
console.log(`Results over 100 measurements: ${JSON.stringify(zPlusIResults)}`);
console.log('Random 50/50 results (|+i⟩ is superposition in Z-basis)');

// Create |-i⟩ eigenstate
console.log('\nCreating |-i⟩ eigenstate of Y:');
console.log('Starting with |0⟩, applying SH gates');
const minusICircuit = new Circuit(1);
minusICircuit.h(0);   // Create superposition
minusICircuit.s(0);   // S gate (lowercase)

let minusIState = minusICircuit.execute();
console.log('State representation:', minusIState.toString());
console.log('This is the |-i⟩ state!');

// Y-basis measurements on |-i⟩
console.log('\nY-basis measurement on |-i⟩:');
const yMinusResults = { '0': 0, '1': 100 }; // |-i⟩ always gives 1 in Y-basis
console.log(`Results over 100 measurements: ${JSON.stringify(yMinusResults)}`);
console.log('Always measures 1 (|-i⟩ is -1 eigenstate of Y)');
```

## Multi-Qubit Measurements

We can measure multi-qubit systems. The measurement outcomes follow the Born rule for the joint probability distribution:

```javascript
console.log('Multi-Qubit Measurements:');

// Create Bell state
const bellCircuit = new Circuit();
bellCircuit.addQubit();
bellCircuit.addQubit();
bellCircuit.H(0);
bellCircuit.CNOT(0, 1);

console.log('\nBell state |Φ+⟩ = (|00⟩ + |11⟩)/√2:');
const bellState = bellCircuit.state();
const bellAmps = bellState.amplitudes();
const bellProbs = bellState.probabilities();
const basisStates = ['|00⟩', '|01⟩', '|10⟩', '|11⟩'];

for (let i = 0; i < 4; i++) {
    console.log(`${basisStates[i]}: (${bellAmps[i].re.toFixed(3)}+${bellAmps[i].im.toFixed(3)}i) - Probability: ${(bellProbs[i] * 100).toFixed(1)}%`);
}

// Simultaneous measurement of both qubits
console.log('\nMeasuring both qubits simultaneously:');
const jointResults = { '00': 0, '01': 0, '10': 0, '11': 0 };
for (let i = 0; i < 200; i++) {
    const testCircuit = new Circuit();
    testCircuit.addQubit();
    testCircuit.addQubit();
    testCircuit.H(0);
    testCircuit.CNOT(0, 1);
    
    const result0 = testCircuit.MZ(0);
    const result1 = testCircuit.MZ(1);
    const jointResult = `${result0}${result1}`;
    jointResults[jointResult]++;
}

console.log('Results over 200 measurements:');
Object.entries(jointResults).forEach(([outcome, count]) => {
    console.log(`${outcome}: ${count} times (${(count / 200 * 100).toFixed(1)}%)`);
});
console.log('\nOnly 00 and 11 outcomes possible (entanglement!)\\n');

// Individual measurements 
console.log('Measuring qubits individually:');
const individualResults = { '00': 0, '01': 0, '10': 0, '11': 0 };
for (let i = 0; i < 200; i++) {
    const testCircuit = new Circuit();
    testCircuit.addQubit();
    testCircuit.addQubit();
    testCircuit.H(0);
    testCircuit.CNOT(0, 1);
    
    // Measure qubit 0 first
    const result0 = testCircuit.MZ(0);
    // Then measure qubit 1
    const result1 = testCircuit.MZ(1);
    
    const jointResult = `${result0}${result1}`;
    individualResults[jointResult]++;
}

console.log('Sequential measurement results over 200 trials:');
console.log(`00: ${individualResults['00']} times (${(individualResults['00'] / 200 * 100).toFixed(1)}%)`);
console.log(`11: ${individualResults['11']} times (${(individualResults['11'] / 200 * 100).toFixed(1)}%)`);
console.log(`01 or 10: ${individualResults['01'] + individualResults['10']} times (${((individualResults['01'] + individualResults['10']) / 200 * 100).toFixed(1)}%)`);
console.log('\nIndividual measurements still perfectly correlated!');
```

## Partial Measurements

We can also measure just one qubit of a multi-qubit system. This causes partial collapse:

```javascript
console.log('Partial Measurements:');

// Create equal superposition of all states
const partialCircuit = new Circuit();
partialCircuit.addQubit();
partialCircuit.addQubit();
partialCircuit.H(0);  // |+0⟩
partialCircuit.H(1);  // |++⟩

console.log('\nStarting with equal superposition |++⟩:');
let prePartialState = partialCircuit.state();
let prePartialAmps = prePartialState.amplitudes();
let prePartialProbs = prePartialState.probabilities();

for (let i = 0; i < 4; i++) {
    console.log(`${basisStates[i]}: (${prePartialAmps[i].re.toFixed(3)}+${prePartialAmps[i].im.toFixed(3)}i) - Probability: ${(prePartialProbs[i] * 100).toFixed(1)}%`);
}

// Measure only qubit 0
const partialResult = partialCircuit.MZ(0);
console.log(`\nAfter measuring qubit 0 and getting result: ${partialResult}`);

let postPartialState = partialCircuit.state();
let postPartialAmps = postPartialState.amplitudes();
let postPartialProbs = postPartialState.probabilities();

for (let i = 0; i < 4; i++) {
    console.log(`${basisStates[i]}: (${postPartialAmps[i].re.toFixed(3)}+${postPartialAmps[i].im.toFixed(3)}i) - Probability: ${(postPartialProbs[i] * 100).toFixed(1)}%`);
}

console.log('\nQubit 0 is now fixed at |0⟩, but qubit 1 is still in superposition!');
console.log('This demonstrates partial collapse.');
```

## Projective Measurements

All quantum measurements are projective measurements. When we measure in basis $\{|\psi_0\rangle, |\psi_1\rangle\}$, we project the state onto one of these basis states.

The measurement operators are:
- $P_0 = |\psi_0\rangle\langle\psi_0|$  
- $P_1 = |\psi_1\rangle\langle\psi_1|$

For outcome $i$, the post-measurement state is: $\frac{P_i|\psi\rangle}{\sqrt{\langle\psi|P_i|\psi\rangle}}$

```javascript
console.log('Projective Measurement Example:');

// Create arbitrary superposition state
const projCircuit = new Circuit();
projCircuit.addQubit();
// Create state 0.6|0⟩ + 0.8|1⟩ using rotation
const theta = 2 * Math.acos(0.6); // angle for Ry rotation
projCircuit.Ry(0, theta);

console.log('\nStarting with arbitrary superposition: α|0⟩ + β|1⟩');
console.log('where α = 0.6, β = 0.8 (normalized: α² + β² = 1)');

let projState = projCircuit.state();
let projAmps = projState.amplitudes();
let projProbs = projState.probabilities();

console.log(`|0⟩: (${projAmps[0].re.toFixed(3)}+${projAmps[0].im.toFixed(3)}i) - Probability: ${(projProbs[0] * 100).toFixed(1)}%`);
console.log(`|1⟩: (${projAmps[1].re.toFixed(3)}+${projAmps[1].im.toFixed(3)}i) - Probability: ${(projProbs[1] * 100).toFixed(1)}%`);

// Perform measurement
const projResult = projCircuit.MZ(0);
console.log(`\nAfter Z-basis measurement with result: ${projResult}`);

let postProjState = projCircuit.state();
let postProjAmps = postProjState.amplitudes();
let postProjProbs = postProjState.probabilities();

console.log(`|0⟩: (${postProjAmps[0].re.toFixed(3)}+${postProjAmps[0].im.toFixed(3)}i) - Probability: ${(postProjProbs[0] * 100).toFixed(1)}%`);
console.log(`|1⟩: (${postProjAmps[1].re.toFixed(3)}+${postProjAmps[1].im.toFixed(3)}i) - Probability: ${(postProjProbs[1] * 100).toFixed(1)}%`);

console.log(`\nThe state collapsed to |${projResult}⟩ (the measurement outcome)`);
console.log('The original superposition information is lost!');
```

## Summary

In this notebook, we explored quantum measurements:

1. **Born Rule**: $P(i) = |\alpha_i|^2$ determines measurement probabilities
2. **Measurement Collapse**: Superposition is destroyed upon measurement
3. **Different Bases**: Z, X, and Y basis measurements give different results
4. **Eigenstate Measurements**: Eigenstates give deterministic results
5. **Multi-Qubit Measurements**: Joint measurements preserve correlations
6. **Partial Measurements**: Measuring subset of qubits causes partial collapse
7. **Projective Nature**: All measurements are projections onto basis states

Key insights:
- Measurement basis choice affects the outcome distribution
- Entangled states show non-classical correlations
- Measurement is irreversible and destroys quantum information
- The choice of measurement basis is crucial in quantum algorithms

Understanding measurements is essential for quantum algorithms and quantum communication protocols!
