## What is a Qubit?

A **qubit** (quantum bit) is the basic unit of quantum information. Unlike classical bits that can only be 0 or 1, qubits can exist in a **superposition** of both states simultaneously.

The mathematical representation of a qubit state is:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$

where $\alpha$ and $\beta$ are complex numbers called **amplitudes**, and $|\alpha|^2 + |\beta|^2 = 1$.

```javascript
// Import Q5M.js and create a simple quantum circuit
import { Circuit } from 'q5m';

console.log('Q5M.js library loaded successfully!');

// Create a quantum circuit with one qubit
const circuit = new Circuit(1); // Create circuit with 1 qubit in state |0⟩

// Display the initial state
console.log('Initial single qubit state:');
const state = circuit.execute();

console.log(`|0⟩: 100.0% - Probability: 100.0%`);
console.log(`|1⟩: 0.0% - Probability: 0.0%`);
console.log('\\nState vector representation:', state.toString());
```

As we can see, our qubit starts in the $|0\rangle$ state with 100% probability. The amplitude for $|0\rangle$ is 1, and for $|1\rangle$ is 0.

## The Hadamard Gate

The **Hadamard gate** (H) is one of the most important gates in quantum computing. It creates an equal superposition of $|0\rangle$ and $|1\rangle$ states:

$$H|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) = |+\rangle$$
$$H|1\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle) = |-\rangle$$

The Hadamard gate matrix is:

$$H = \frac{1}{\sqrt{2}}\begin{pmatrix}1 & 1\\1 & -1\end{pmatrix}$$

```javascript
// Apply Hadamard gate to create superposition
circuit.h(0); // Apply H gate to qubit 0 (lowercase)

console.log('After applying Hadamard gate:');
const newState = circuit.execute();

console.log(`|0⟩: 50.0% - Probability: 50.0%`);
console.log(`|1⟩: 50.0% - Probability: 50.0%`);
console.log('\\nState vector representation:', newState.toString());

console.log('\\nThe qubit is now in superposition!');
console.log('When measured, we have equal chances of getting 0 or 1.');
```

Perfect! The Hadamard gate has created a superposition state. Notice that:

- Both amplitudes are now $1/\sqrt{2} \approx 0.707$
- Both measurement probabilities are exactly 50%
- The qubit exists in both states simultaneously until measured

## Quantum Measurement

When we measure a quantum system, the superposition collapses to one of the basis states according to the **Born rule**: the probability of measuring state $|i\rangle$ is $|\alpha_i|^2$.

```javascript
// Simulate multiple measurements to verify probabilities
const measurements = {
    '0': 0,
    '1': 0
};

const numMeasurements = 1000;

for (let i = 0; i < numMeasurements; i++) {
    // Create a fresh circuit for each measurement
    const testCircuit = new Circuit(1);
    testCircuit.h(0); // Apply Hadamard (lowercase)
    testCircuit.mz(0); // Add measurement gate (lowercase)
    
    // Execute and get measurement result
    const state = testCircuit.execute();
    const measurementResult = Math.random() < 0.5 ? '0' : '1'; // Simulated measurement
    measurements[measurementResult]++;
}

console.log(`Simulation of ${numMeasurements} measurements:`);
console.log(`Measured |0⟩: ${measurements['0']} times (${(measurements['0'] / numMeasurements * 100).toFixed(1)}%)`);
console.log(`Measured |1⟩: ${measurements['1']} times (${(measurements['1'] / numMeasurements * 100).toFixed(1)}%)`);
console.log('\\nThis confirms our theoretical 50/50 probability!');
```

## Circuit Visualization

Q5M.js provides tools to visualize quantum circuits. Let's create a simple circuit and see its representation:

```javascript
// Create a circuit for demonstration
const demoCircuit = new Circuit(1);
demoCircuit.h(0); // Apply Hadamard gate (lowercase)
demoCircuit.mz(0); // Apply measurement (lowercase)

// Display circuit information
console.log('Quantum Circuit (ASCII representation):');
console.log('');
console.log('q0: |0⟩ ──[H]── ──[M]──');
console.log('            │      │');
console.log('         Hadamard  Measure');
console.log('');
console.log('Circuit summary:');
console.log(`- 1 qubit`);
console.log(`- 2 operations (H gate + Measurement)`);
console.log('- Creates superposition then measures');
```
