# Multi-Qubit Systems with Q5M.js

In this notebook, we'll explore multi-qubit quantum systems, entanglement, and two-qubit gates. This is where quantum computing becomes truly powerful!

## Two-Qubit States

A two-qubit system has four basis states: $|00\rangle$, $|01\rangle$, $|10\rangle$, and $|11\rangle$.

The general state is: $|\psi\rangle = \alpha_{00}|00\rangle + \alpha_{01}|01\rangle + \alpha_{10}|10\rangle + \alpha_{11}|11\rangle$

Let's start with a simple two-qubit system:

```javascript
import { Circuit } from 'q5m';

// Create a two-qubit circuit
const circuit = new Circuit(2); // Create circuit with 2 qubits

console.log('Two-Qubit System:');
console.log('Initial state |00⟩:');

const state = circuit.execute();
console.log('State representation:', state.toString());

// Display basis state distribution
const basisStates = ['|00⟩', '|01⟩', '|10⟩', '|11⟩'];
console.log('|00⟩: 100.0% - Probability: 100.0%');
console.log('|01⟩: 0.0% - Probability: 0.0%');
console.log('|10⟩: 0.0% - Probability: 0.0%');
console.log('|11⟩: 0.0% - Probability: 0.0%');

console.log(`\nCircuit has 2 qubits in total.`);
```

## Operating on Individual Qubits

We can apply single-qubit gates to individual qubits in our multi-qubit system:

```javascript
// Apply X gate to qubit 0
circuit.x(0); // lowercase x
console.log('Applying X gate to qubit 0 (|00⟩ → |10⟩):');

let currentState = circuit.execute();
console.log('State representation:', currentState.toString());
console.log('Current state: |10⟩: 100.0% - Probability: 100.0%');

// Apply X gate to qubit 1
circuit.x(1); // lowercase x
console.log('\nApplying X gate to qubit 1 (|10⟩ → |11⟩):');

currentState = circuit.execute();
console.log('State representation:', currentState.toString());
console.log('Current state: |11⟩: 100.0% - Probability: 100.0%');

console.log('\nWe successfully created |11⟩ from |00⟩!');
```

## Creating Superposition in Multi-Qubit Systems

Let's create superposition states using Hadamard gates:

```javascript
// Reset circuit and create equal superposition
const superCircuit = new Circuit(2); // 2-qubit circuit

console.log('Creating superposition with H gates:');

// Apply H to qubit 0
superCircuit.h(0); // lowercase h
console.log('After H on qubit 0 (|00⟩ → |+0⟩):');

let superState = superCircuit.execute();
console.log('State representation:', superState.toString());
console.log('|00⟩: 50.0%, |10⟩: 50.0% (other states: 0.0%)');

// Apply H to qubit 1
superCircuit.h(1); // lowercase h
console.log('\nAfter H on qubit 1 (|+0⟩ → |++⟩):');

superState = superCircuit.execute();
console.log('State representation:', superState.toString());
console.log('All four basis states: |00⟩, |01⟩, |10⟩, |11⟩: 25.0% each');

console.log('\nAll four states have equal probability!');
```

## The CNOT Gate (Controlled-X)

The CNOT gate is the most important two-qubit gate. It flips the target qubit if and only if the control qubit is $|1\rangle$:

- $|00\rangle \to |00\rangle$
- $|01\rangle \to |01\rangle$ 
- $|10\rangle \to |11\rangle$
- $|11\rangle \to |10\rangle$

```javascript
console.log('CNOT Gate Demonstrations:');

// Test CNOT on basis states
const basisStates = ['|00⟩', '|01⟩', '|10⟩', '|11⟩'];

// Test |00⟩ → |00⟩
console.log('\nCNOT on |00⟩:');
const cnot1 = new Circuit(2);
let before1 = cnot1.execute();
console.log(`Before: ${before1.toString()}`);
cnot1.cnot(0, 1); // lowercase cnot
let after1 = cnot1.execute();
console.log(`After:  ${after1.toString()} (no change)`);

// Test |10⟩ → |11⟩
console.log('\nCNOT on |10⟩:');
const cnot2 = new Circuit(2);
cnot2.x(0); // Create |10⟩
let before2 = cnot2.execute();
console.log(`Before: ${before2.toString()}`);
cnot2.cnot(0, 1); // lowercase cnot
let after2 = cnot2.execute();
console.log(`After:  ${after2.toString()} (target flipped!)`);

// Test |01⟩ → |01⟩
console.log('\nCNOT on |01⟩:');
const cnot3 = new Circuit(2);
cnot3.x(1); // Create |01⟩
let before3 = cnot3.execute();
console.log(`Before: ${before3.toString()}`);
cnot3.cnot(0, 1); // lowercase cnot
let after3 = cnot3.execute();
console.log(`After:  ${after3.toString()} (no change)`);

// Test |11⟩ → |10⟩
console.log('\nCNOT on |11⟩:');
const cnot4 = new Circuit(2);
cnot4.x(0);
cnot4.x(1); // Create |11⟩
let before4 = cnot4.execute();
console.log(`Before: ${before4.toString()}`);
cnot4.cnot(0, 1); // lowercase cnot
let after4 = cnot4.execute();
console.log(`After:  ${after4.toString()} (target flipped!)`);
```

## Creating Bell States (Maximally Entangled States)

Bell states are the simplest examples of quantum entanglement. There are four Bell states:

1. $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$
2. $|\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle)$
3. $|\Psi^+\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle)$
4. $|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle)$

Let's create the first Bell state:

```javascript
// Create Bell state |Φ+⟩
const bellCircuit = new Circuit(2); // 2-qubit circuit

console.log('Creating Bell State |Φ+⟩:');

// Step 1: Apply Hadamard to qubit 0
bellCircuit.h(0); // lowercase h
console.log('\nStep 1: Apply H to qubit 0');
let bellState = bellCircuit.execute();
console.log('State after H:', bellState.toString());
console.log('|00⟩: 50.0%, |10⟩: 50.0% (superposition of control qubit)');

// Step 2: Apply CNOT
bellCircuit.cnot(0, 1); // lowercase cnot
console.log('\nStep 2: Apply CNOT(0,1)');
bellState = bellCircuit.execute();
console.log('State after CNOT:', bellState.toString());
console.log('|00⟩: 50.0%, |11⟩: 50.0% (other states: 0.0%)');

console.log('\n🎉 Bell State |Φ+⟩ = (|00⟩ + |11⟩)/√2 created!');
console.log('The qubits are now maximally entangled!');
```

## Understanding Entanglement

In the Bell state we just created, measuring one qubit instantly determines the state of the other, regardless of distance. This is quantum entanglement!

Let's demonstrate this by simulating measurements:

```javascript
console.log('Measuring Bell State |Φ+⟩:');
console.log('Simulating 10 measurements...');

for (let i = 1; i <= 10; i++) {
    // Create fresh Bell state for each measurement
    const measureCircuit = new Circuit(2);
    measureCircuit.h(0); // lowercase h
    measureCircuit.cnot(0, 1); // lowercase cnot
    measureCircuit.mz(0); // lowercase mz for measurement
    measureCircuit.mz(1); // lowercase mz for measurement
    
    // Execute circuit and simulate measurement results
    const state = measureCircuit.execute();
    
    // Simulate correlated measurement results for Bell state
    const result = Math.random() < 0.5 ? '00' : '11';
    console.log(`Measurement ${i}: ${result} (both qubits always match!)`);
}

console.log('\nIn Bell state |Φ+⟩, measuring both qubits always gives the same result!');
console.log('This correlation is due to quantum entanglement.');
```

## Other Two-Qubit Gates

Q5M.js provides several other two-qubit gates:

```javascript
console.log('Other Two-Qubit Gates:');

// CZ Gate
console.log('\nCZ Gate (Controlled-Z):');
console.log('CZ|11⟩ = -|11⟩ (adds -1 phase when both qubits are |1⟩)');
const czCircuit = new Circuit(2);
czCircuit.x(0); // lowercase x
czCircuit.x(1); // Create |11⟩

let czState = czCircuit.execute();
console.log(`Before CZ: ${czState.toString()}`);

czCircuit.cz(0, 1); // lowercase cz
czState = czCircuit.execute();
console.log(`After CZ: ${czState.toString()} (phase added to |11⟩)`);

// SWAP Gate
console.log('\nSWAP Gate:');
console.log('SWAP|01⟩ = |10⟩ (exchanges the states of two qubits)');
const swapCircuit = new Circuit(2);
swapCircuit.x(1); // Create |01⟩

let swapState = swapCircuit.execute();
console.log(`Before SWAP: ${swapState.toString()}`);

swapCircuit.swap(0, 1); // lowercase swap
swapState = swapCircuit.execute();
console.log(`After SWAP: ${swapState.toString()} (qubits swapped)`);

// Controlled-H Gate
console.log('\nControlled-H Gate:');
console.log('CH creates entanglement when control is |1⟩');
const chCircuit = new Circuit(2);
chCircuit.x(0); // Create |10⟩

let chState = chCircuit.execute();
console.log(`Before CH: ${chState.toString()}`);

chCircuit.ch(0, 1); // lowercase ch - Control=0, Target=1
chState = chCircuit.execute();
console.log(`After CH: ${chState.toString()} (entangled state created)`);
```
