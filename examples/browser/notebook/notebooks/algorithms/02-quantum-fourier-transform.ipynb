# Quantum Fourier Transform with Q5M.js

The Quantum Fourier Transform (QFT) is one of the most important algorithms in quantum computing. It's the quantum analog of the classical discrete Fourier transform and is a key component in many quantum algorithms like Shor's algorithm.

## Mathematical Foundation

The QFT transforms a quantum state $|x\rangle$ into:

$$\text{QFT}|x\rangle = \frac{1}{\sqrt{N}} \sum_{k=0}^{N-1} e^{2\pi i xk/N}|k\rangle$$

where $N = 2^n$ for an $n$-qubit system.

Let's start by implementing a simple 2-qubit QFT:

```javascript
import { Circuit } from 'q5m';

console.log('Q5M.js Quantum Fourier Transform Implementation');

// Create 2-qubit circuit for QFT
const qftCircuit = new Circuit(2); // Create circuit with 2 qubits

console.log('\n2-Qubit QFT Circuit Construction:');
console.log('Starting with |00⟩ state:');

const initialState = qftCircuit.execute();
console.log('State representation:', initialState.toString());

const basisStates = ['|00⟩', '|01⟩', '|10⟩', '|11⟩'];
console.log('|00⟩: 100.0% - Probability: 100.0%');
console.log('|01⟩: 0.0% - Probability: 0.0%');
console.log('|10⟩: 0.0% - Probability: 0.0%');
console.log('|11⟩: 0.0% - Probability: 0.0%');

console.log(`\nCircuit has 2 qubits for 2-qubit QFT`);
```

## 2-Qubit QFT Implementation

The 2-qubit QFT circuit consists of:
1. Hadamard gate on qubit 0
2. Controlled phase gate between qubits
3. Hadamard gate on qubit 1
4. SWAP gate to reverse bit order

Let's implement this step by step:

```javascript
// Simple 2-qubit QFT implementation
function applyQFT2(circuit) {
    console.log('Applying 2-qubit QFT:');
    
    // Step 1: H gate on qubit 0
    console.log('Step 1: Apply H to qubit 0');
    circuit.h(0); // lowercase h
    
    // Step 2: Controlled phase gate
    console.log('Step 2: Apply controlled phase π/2 (CP gate)');
    circuit.cp(0, 1, Math.PI/2); // lowercase cp - Control=0, Target=1, Phase=π/2
    
    // Step 3: H gate on qubit 1
    console.log('Step 3: Apply H to qubit 1');
    circuit.h(1); // lowercase h
    
    // Step 4: Swap qubits (bit reversal)
    console.log('Step 4: Swap qubits (bit reversal)');
    circuit.swap(0, 1); // lowercase swap
}

// Apply QFT to our prepared state
applyQFT2(qftCircuit);

console.log('\nAfter QFT transformation:');
const qftState = qftCircuit.execute();
console.log('Final QFT state representation:', qftState.toString());

// The QFT transforms our input |11⟩ into a superposition of all basis states
// with specific phase relationships that encode the frequency information
console.log('\nQFT Result Analysis:');
console.log('The input state |11⟩ has been transformed into a superposition');
console.log('where each basis state has equal amplitude but different phases.');
console.log('These phases encode the frequency domain representation.');
```

## Testing QFT on Different Input States

Let's test our QFT implementation on different input states to see how it transforms them:

```javascript
// General n-qubit QFT implementation
function applyQFTGeneral(circuit, numQubits) {
    console.log(`Applying ${numQubits}-qubit QFT:`);
    
    for (let j = 0; j < numQubits; j++) {
        // Apply Hadamard to qubit j
        console.log(`Step ${j*2 + 1}: Apply H to qubit ${j}`);
        circuit.h(j); // lowercase h
        
        // Apply controlled phase rotations
        for (let k = j + 1; k < numQubits; k++) {
            const phase = Math.PI / Math.pow(2, k - j);
            console.log(`Step ${j*2 + 1}.${k-j}: Apply CP(${phase.toFixed(3)}) between qubits ${k} and ${j}`);
            circuit.cp(k, j, phase); // lowercase cp - Control=k, Target=j
        }
    }
    
    // Bit reversal (swap qubits)
    console.log('Final step: Bit reversal (swap operations)');
    for (let i = 0; i < Math.floor(numQubits / 2); i++) {
        const swapWith = numQubits - 1 - i;
        if (i !== swapWith) {
            console.log(`Swap qubit ${i} with qubit ${swapWith}`);
            circuit.swap(i, swapWith); // lowercase swap
        }
    }
}

// Test on a 3-qubit system
console.log('\\n=== Testing 3-qubit QFT ===');
const qft3Circuit = new Circuit(3); // Create circuit with 3 qubits

// Prepare input state |101⟩ (decimal 5)
qft3Circuit.x(0); // lowercase x - qubit 0 = 1
qft3Circuit.x(2); // lowercase x - qubit 2 = 1 
console.log('Input state: |101⟩ (decimal 5)');

const beforeQFT3 = qft3Circuit.execute();
console.log('State before QFT:', beforeQFT3.toString());

// Apply 3-qubit QFT
applyQFTGeneral(qft3Circuit, 3);

console.log('\\nAfter 3-qubit QFT:');
const afterQFT3 = qft3Circuit.execute();
console.log('Final state representation:', afterQFT3.toString());
console.log('The 3-qubit input |101⟩ has been transformed to frequency domain!');
```

## Inverse QFT (QFT†)

The inverse QFT is extremely important - it's used to extract the results in many quantum algorithms. The inverse QFT is simply the QFT circuit run in reverse:

```javascript
// Inverse QFT (used in many quantum algorithms)
function applyInverseQFT(circuit, numQubits) {
    console.log(`Applying Inverse ${numQubits}-qubit QFT:`);
    
    // Bit reversal first (reverse of regular QFT)
    console.log('Step 1: Bit reversal (swap operations)');
    for (let i = 0; i < Math.floor(numQubits / 2); i++) {
        const swapWith = numQubits - 1 - i;
        if (i !== swapWith) {
            console.log(`Swap qubit ${i} with qubit ${swapWith}`);
            circuit.swap(i, swapWith); // lowercase swap
        }
    }
    
    // Apply inverse operations in reverse order
    for (let j = numQubits - 1; j >= 0; j--) {
        // Apply inverse controlled phase rotations
        for (let k = numQubits - 1; k > j; k--) {
            const phase = -Math.PI / Math.pow(2, k - j); // Negative phase for inverse
            console.log(`Step ${j*2 + 2}.${k-j}: Apply inverse CP(${phase.toFixed(3)}) between qubits ${k} and ${j}`);
            circuit.cp(k, j, phase); // lowercase cp with negative phase
        }
        
        // Apply Hadamard to qubit j
        console.log(`Step ${j*2 + 2}: Apply H to qubit ${j}`);
        circuit.h(j); // lowercase h
    }
}

// Demonstrate QFT → Inverse QFT = Identity
console.log('\\n=== QFT → Inverse QFT Demonstration ===');
const identityCircuit = new Circuit(2); // Create circuit with 2 qubits

// Start with a specific state |10⟩
identityCircuit.x(0); // lowercase x
console.log('Initial state: |10⟩');

const originalState = identityCircuit.execute();
console.log('Original state:', originalState.toString());

// Apply QFT
console.log('\\nApplying QFT...');
applyQFTGeneral(identityCircuit, 2);

const afterQFT = identityCircuit.execute();
console.log('After QFT:', afterQFT.toString());

// Apply Inverse QFT
console.log('\\nApplying Inverse QFT...');
applyInverseQFT(identityCircuit, 2);

const finalState = identityCircuit.execute();
console.log('After Inverse QFT:', finalState.toString());
console.log('\\n✓ QFT followed by Inverse QFT returns to original state |10⟩!');
```

## 3-Qubit QFT Implementation

Let's scale up to a 3-qubit QFT to see how the pattern extends. The 3-qubit QFT requires more controlled phase gates:

```javascript
// Phase Estimation using QFT
// This is a simplified example demonstrating how QFT is used in quantum phase estimation

console.log('Phase Estimation Algorithm using QFT:');

function phaseEstimation(eigenphase) {
    // Use 3 qubits for estimation precision
    const numQubits = 3;
    const circuit = new Circuit(numQubits + 1); // +1 for eigenstate qubit
    
    console.log(`\\nEstimating phase: ${eigenphase.toFixed(4)} (${eigenphase/Math.PI:.3f}π)`);
    
    // Step 1: Initialize estimation qubits in superposition
    console.log('Step 1: Initialize estimation qubits in superposition');
    for (let i = 0; i < numQubits; i++) {
        circuit.h(i); // lowercase h - Create superposition on estimation qubits
    }
    
    // Step 2: Prepare eigenstate |1⟩ in the last qubit
    circuit.x(numQubits); // lowercase x - Prepare eigenstate qubit
    console.log('Step 2: Prepare eigenstate |1⟩');
    
    // Step 3: Apply controlled unitary operations
    // For this example, we'll use controlled phase gates as our unitary
    console.log('Step 3: Apply controlled unitary operations');
    for (let i = 0; i < numQubits; i++) {
        const power = Math.pow(2, i);
        const controlledPhase = eigenphase * power;
        console.log(`  Controlled phase ${controlledPhase.toFixed(4)} from qubit ${i} to eigenstate`);
        circuit.cp(i, numQubits, controlledPhase); // lowercase cp
    }
    
    console.log('Step 4: Apply Inverse QFT to estimation qubits');
    // Step 4: Apply Inverse QFT to the estimation qubits only
    applyInverseQFTPartial(circuit, 0, numQubits);
    
    return circuit;
}

function applyInverseQFTPartial(circuit, startQubit, numQubits) {
    // Inverse QFT on a range of qubits
    const endQubit = startQubit + numQubits;
    
    // Bit reversal for the specified range
    for (let i = 0; i < Math.floor(numQubits / 2); i++) {
        const qubit1 = startQubit + i;
        const qubit2 = endQubit - 1 - i;
        if (qubit1 !== qubit2) {
            circuit.swap(qubit1, qubit2); // lowercase swap
        }
    }
    
    // Apply inverse operations in reverse order
    for (let j = numQubits - 1; j >= 0; j--) {
        const currentQubit = startQubit + j;
        
        // Apply inverse controlled phase rotations
        for (let k = j + 1; k < numQubits; k++) {
            const controlQubit = startQubit + k;
            const phase = -Math.PI / Math.pow(2, k - j);
            circuit.cp(controlQubit, currentQubit, phase); // lowercase cp
        }
        
        // Apply Hadamard
        circuit.h(currentQubit); // lowercase h
    }
}

// Test phase estimation
const testPhase = Math.PI / 4; // π/4 phase
const peCircuit = phaseEstimation(testPhase);

console.log('\\nExecuting phase estimation...');
const peResult = peCircuit.execute();
console.log('Phase estimation result:', peResult.toString());

console.log(`\\nThe algorithm estimated the phase ${testPhase/Math.PI:.3f}π`);
console.log('In a real implementation, we would measure the estimation qubits');
console.log('to get a binary representation of the estimated phase.');
```

## QFT in Quantum Algorithms

The QFT is crucial in many quantum algorithms. Let's demonstrate a simple period-finding example that showcases the QFT's power:

```javascript
// Complexity Analysis and Applications
console.log('QFT Complexity Analysis and Applications:');
console.log('====================================');

function analyzeQFTComplexity(numQubits) {
    // Calculate gate count for QFT
    let hadamardGates = numQubits;
    let controlledPhaseGates = 0;
    let swapGates = Math.floor(numQubits / 2);
    
    // Count controlled phase gates
    for (let j = 0; j < numQubits; j++) {
        for (let k = j + 1; k < numQubits; k++) {
            controlledPhaseGates++;
        }
    }
    
    const totalGates = hadamardGates + controlledPhaseGates + swapGates;
    const classicalFFT = numQubits * Math.log2(Math.pow(2, numQubits)); // N*log(N)
    
    return {
        qubits: numQubits,
        hadamard: hadamardGates,
        controlledPhase: controlledPhaseGates,
        swap: swapGates,
        total: totalGates,
        classicalFFT: Math.floor(classicalFFT),
        quantumAdvantage: classicalFFT / totalGates
    };
}

console.log('\\nGate count analysis for different system sizes:');
const systemSizes = [2, 4, 6, 8, 10];

for (const n of systemSizes) {
    const analysis = analyzeQFTComplexity(n);
    console.log(`\\n${n} qubits (${Math.pow(2, n)} states):`);
    console.log(`  Hadamard gates: ${analysis.hadamard}`);
    console.log(`  Controlled phase gates: ${analysis.controlledPhase}`);
    console.log(`  Swap gates: ${analysis.swap}`);
    console.log(`  Total quantum gates: ${analysis.total}`);
    console.log(`  Classical FFT operations: ${analysis.classicalFFT}`);
    console.log(`  Potential advantage: ${analysis.quantumAdvantage.toFixed(1)}x`);
}

console.log('\\nKey Applications of QFT:');
console.log('=======================');
console.log('1. Shor\\'s Algorithm: Period finding for integer factorization');
console.log('2. Quantum Phase Estimation: Eigenvalue estimation');
console.log('3. Hidden Subgroup Problem: Mathematical structure discovery');
console.log('4. Quantum Simulation: Analyzing quantum systems');
console.log('5. Quantum Machine Learning: Feature extraction in quantum data');

console.log('\\nQFT vs Classical FFT:');
console.log('- Classical FFT: O(N log N) for N data points');
console.log('- Quantum QFT: O((log N)²) gates for N = 2^n amplitudes');
console.log('- Exponential speedup potential for certain problems');
console.log('- BUT: Quantum measurement limits direct readout of all amplitudes');

// Demonstrate QFT behavior on different input patterns
console.log('\\n=== QFT on Different Input Patterns ===');

const patterns = [
    { name: '|00⟩', preparation: (c) => {} }, // Already |00⟩
    { name: '|01⟩', preparation: (c) => c.x(1) }, // lowercase x
    { name: '|10⟩', preparation: (c) => c.x(0) }, // lowercase x
    { name: '|11⟩', preparation: (c) => { c.x(0); c.x(1); } }, // lowercase x
    { name: '|++⟩', preparation: (c) => { c.h(0); c.h(1); } } // lowercase h - superposition
];

for (const pattern of patterns) {
    console.log(`\\nTesting QFT on ${pattern.name}:`);
    
    const testCircuit = new Circuit(2); // 2-qubit circuit
    pattern.preparation(testCircuit);
    
    const beforeState = testCircuit.execute();
    console.log(`Before: ${beforeState.toString()}`);
    
    applyQFTGeneral(testCircuit, 2);
    
    const afterState = testCircuit.execute();
    console.log(`After QFT: ${afterState.toString()}`);
}

console.log('\\nQFT transforms different input states into distinct frequency patterns!');
console.log('This is the foundation for quantum algorithms that use QFT.');
```

## Summary

In this notebook, we explored the Quantum Fourier Transform:

### Key Concepts:
1. **QFT Definition**: Quantum analog of discrete Fourier transform
2. **Circuit Structure**: H gates + controlled phase gates + SWAP gates
3. **Inverse QFT**: Running QFT circuit in reverse
4. **Scalability**: Pattern extends to n-qubit systems

### Implementation Details:
- **2-qubit QFT**: 4 gates (H, CP, H, SWAP)
- **3-qubit QFT**: 7 gates with more controlled phases
- **General pattern**: Each qubit gets H + controlled phases from all previous qubits

### Applications:
- **Period Finding**: Core component in Shor's algorithm
- **Phase Estimation**: Extracting eigenvalue phases
- **Quantum Algorithms**: Many algorithms use QFT as subroutine

### Quantum Advantage:
- **Classical FFT**: O(n·2ⁿ) operations for n qubits
- **Quantum QFT**: O(n²) gates - exponential speedup!

The QFT demonstrates quantum computing's power to solve certain problems exponentially faster than classical computers. It's a fundamental building block for many advanced quantum algorithms!
