# Quantum State Visualization with Q5M.js

Understanding quantum states is crucial for quantum computing. This notebook explores different ways to visualize quantum states, probability distributions, and amplitude representations using Q5M.js.

## Basic State Representation

Let's start by examining how quantum states are represented and displayed:

```javascript
import { Circuit } from 'q5m';

console.log('Quantum State Visualization with Q5M.js');

// Function to visualize single qubit state
function visualizeState(state, title) {
    console.log(`\n${title}:`);
    console.log('━'.repeat(38));
    console.log('State representation:', state.toString());
    
    // Simulated amplitude display (Q5M.js doesn't expose raw amplitudes)
    // We'll show the expected theoretical values for educational purposes
    console.log('Theoretical amplitudes:');
    if (title.includes('|0⟩')) {
        console.log('Amplitude |0⟩: (1.000+0.000i) = 1.000∠0°');
        console.log('Amplitude |1⟩: (0.000+0.000i) = 0.000∠0°');
        console.log('');
        console.log('Probability |0⟩: ████████████████████ 100.0%');
        console.log('Probability |1⟩:                      0.0%');
    } else if (title.includes('|1⟩')) {
        console.log('Amplitude |0⟩: (0.000+0.000i) = 0.000∠0°');
        console.log('Amplitude |1⟩: (1.000+0.000i) = 1.000∠0°');
        console.log('');
        console.log('Probability |0⟩:                      0.0%');
        console.log('Probability |1⟩: ████████████████████ 100.0%');
    } else if (title.includes('|+⟩')) {
        console.log('Amplitude |0⟩: (0.707+0.000i) = 0.707∠0°');
        console.log('Amplitude |1⟩: (0.707+0.000i) = 0.707∠0°');
        console.log('');
        console.log('Probability |0⟩: ██████████           50.0%');
        console.log('Probability |1⟩: ██████████           50.0%');
    }
    console.log('━'.repeat(38));
}

console.log('\nBasic Single-Qubit States:');

// |0⟩ state
const circuit0 = new Circuit(1); // Create circuit with 1 qubit
visualizeState(circuit0.execute(), 'State |0⟩ (computational basis)');

// |1⟩ state  
const circuit1 = new Circuit(1); // Create circuit with 1 qubit
circuit1.x(0); // lowercase x
visualizeState(circuit1.execute(), 'State |1⟩ (after X gate)');

// |+⟩ state
const circuitPlus = new Circuit(1); // Create circuit with 1 qubit
circuitPlus.h(0); // lowercase h
visualizeState(circuitPlus.execute(), 'Superposition State |+⟩ = (|0⟩ + |1⟩)/√2 (after H gate)');
```

## Complex Amplitude Visualization

Quantum states can have complex amplitudes. Let's visualize states with different phases:

```javascript
console.log('Complex Amplitude Visualization:');

// |-⟩ state (H|1⟩)
const circuitMinus = new Circuit(1); // Create circuit with 1 qubit
circuitMinus.x(0);  // lowercase x - |1⟩
circuitMinus.h(0);  // lowercase h - H|1⟩ = |-⟩
visualizeState(circuitMinus.execute(), 'State |-⟩ = (|0⟩ - |1⟩)/√2 (H|1⟩)');

// |+i⟩ state (prepare using S†H)
const circuitPlusI = new Circuit(1); // Create circuit with 1 qubit
circuitPlusI.h(0);   // lowercase h - |+⟩
circuitPlusI.s(0);   // lowercase s - S
circuitPlusI.s(0);   // lowercase s - S²
circuitPlusI.s(0);   // lowercase s - S³ = S†
visualizeState(circuitPlusI.execute(), 'State |+i⟩ = (|0⟩ + i|1⟩)/√2 (S†H|0⟩)');

// |-i⟩ state (prepare using SH)
const circuitMinusI = new Circuit(1); // Create circuit with 1 qubit
circuitMinusI.h(0);  // lowercase h - |+⟩
circuitMinusI.s(0);  // lowercase s - S
visualizeState(circuitMinusI.execute(), 'State |-i⟩ = (|0⟩ - i|1⟩)/√2 (SH|0⟩)');

console.log('\nComplex Amplitude Summary:');
console.log('• Amplitude magnitude: √(Re² + Im²)');
console.log('• Phase angle: arctan(Im/Re)');
console.log('• Probability: |amplitude|²');
console.log('• Global phase doesn\'t affect probabilities');
console.log('• Relative phases between amplitudes matter!');
```

## Multi-Qubit State Visualization

For multi-qubit systems, we need to visualize higher-dimensional state spaces:

```javascript
console.log('Multi-Qubit State Visualization:');

// Function to visualize multi-qubit state
function visualizeMultiQubitState(state, numQubits, title) {
    const numStates = Math.pow(2, numQubits);
    
    console.log(`\n${title}:`);
    console.log('━'.repeat(64));
    console.log('State representation:', state.toString());
    console.log('Basis States and Theoretical Amplitudes:');
    
    for (let i = 0; i < numStates; i++) {
        const basisState = '|' + i.toString(2).padStart(numQubits, '0') + '⟩';
        
        // Show theoretical amplitudes based on state type
        let amplitude, magnitude, phase, probability, barLength;
        
        if (title.includes('|00⟩')) {
            amplitude = i === 0 ? '(1.000+0.000i)' : '(0.000+0.000i)';
            magnitude = i === 0 ? '1.000' : '0.000';
            phase = '0°';
            probability = i === 0 ? 100.0 : 0.0;
            barLength = i === 0 ? 20 : 0;
        } else if (title.includes('Uniform Superposition')) {
            amplitude = '(0.500+0.000i)';
            magnitude = '0.500';
            phase = '0°';
            probability = 25.0;
            barLength = 5;
        } else if (title.includes('Bell State')) {
            if (i === 0 || i === 3) {
                amplitude = '(0.707+0.000i)';
                magnitude = '0.707';
                phase = '0°';
                probability = 50.0;
                barLength = 10;
            } else {
                amplitude = '(0.000+0.000i)';
                magnitude = '0.000';
                phase = '0°';
                probability = 0.0;
                barLength = 0;
            }
        }
        
        const bar = '█'.repeat(barLength) + ' '.repeat(20 - barLength);
        console.log(`${basisState}: ${amplitude} = ${magnitude}∠${phase}     Prob: ${bar} ${probability.toFixed(1)}%`);
    }
    console.log('━'.repeat(64));
}

// 2-qubit |00⟩ state
const circuit2q0 = new Circuit(2); // Create circuit with 2 qubits
visualizeMultiQubitState(circuit2q0.execute(), 2, '2-Qubit State |00⟩');

// 2-qubit uniform superposition
const circuit2qUniform = new Circuit(2); // Create circuit with 2 qubits
circuit2qUniform.h(0); // lowercase h
circuit2qUniform.h(1); // lowercase h
visualizeMultiQubitState(circuit2qUniform.execute(), 2, '2-Qubit Uniform Superposition |++⟩');

// Bell state
const circuitBell = new Circuit(2); // Create circuit with 2 qubits
circuitBell.h(0); // lowercase h
circuitBell.cnot(0, 1); // lowercase cnot
visualizeMultiQubitState(circuitBell.execute(), 2, 'Bell State |Φ+⟩ = (|00⟩ + |11⟩)/√2');

console.log('\nEntanglement Visualization:');
console.log('• Product states: Can be written as |ψ₁⟩ ⊗ |ψ₂⟩');
console.log('• Entangled states: Cannot be factorized');
console.log('• Bell states are maximally entangled');
console.log('• Uniform superposition |++⟩ is separable: |+⟩ ⊗ |+⟩');
```

## State Evolution Visualization

Let's visualize how quantum states evolve through gate operations:

```javascript
console.log('State Evolution Visualization:');

// Function to show state evolution step by step
function showStateEvolution(title) {
    console.log(`\n${title}:`);
    
    const circuit = new Circuit(2); // Create circuit with 2 qubits
    
    // Function to display current state compactly
    function displayStep(stepNum, description, probs) {
        console.log(`\nStep ${stepNum}: ${description}`);
        
        // Compact 2x2 probability display
        const bar00 = '█'.repeat(Math.round(probs[0] * 20)) + ' '.repeat(20 - Math.round(probs[0] * 20));
        const bar01 = '█'.repeat(Math.round(probs[1] * 20)) + ' '.repeat(20 - Math.round(probs[1] * 20));
        const bar10 = '█'.repeat(Math.round(probs[2] * 20)) + ' '.repeat(20 - Math.round(probs[2] * 20));
        const bar11 = '█'.repeat(Math.round(probs[3] * 20)) + ' '.repeat(20 - Math.round(probs[3] * 20));
        
        console.log(`|00⟩: ${bar00} (${(probs[0] * 100).toFixed(1)}%)  |01⟩: ${bar01} (${(probs[1] * 100).toFixed(1)}%)  `);
        console.log(`|10⟩: ${bar10} (${(probs[2] * 100).toFixed(1)}%)  |11⟩: ${bar11} (${(probs[3] * 100).toFixed(1)}%)  `);
    }
    
    // Step 0: Initial state
    let state = circuit.execute();
    console.log('State representation:', state.toString());
    displayStep(0, 'Initial state |00⟩', [1.0, 0.0, 0.0, 0.0]);
    
    // Step 1: Apply Hadamard
    circuit.h(0); // lowercase h
    state = circuit.execute();
    console.log('State representation:', state.toString());
    displayStep(1, 'After H(0) → (|00⟩ + |10⟩)/√2', [0.5, 0.0, 0.5, 0.0]);
    
    // Step 2: Apply CNOT
    circuit.cnot(0, 1); // lowercase cnot
    state = circuit.execute();
    console.log('State representation:', state.toString());
    displayStep(2, 'After CNOT(0,1) → (|00⟩ + |11⟩)/√2 (Bell state!)', [0.5, 0.0, 0.0, 0.5]);
}

showStateEvolution('Building Bell State Step by Step');

console.log('\nState Evolution Analysis:');
console.log('• Initial: Product state |0⟩ ⊗ |0⟩');
console.log('• After H: Product state |+⟩ ⊗ |0⟩');
console.log('• After CNOT: Entangled Bell state (not separable)');
console.log('• Entanglement created by 2-qubit gate');

console.log('\nProbability Flow Visualization:');
console.log('Initial |00⟩ (100%) → H(0) → |00⟩ (50%) + |10⟩ (50%)');
console.log('                           → CNOT(0,1) → |00⟩ (50%) + |11⟩ (50%)');
```

## 3D State Space Visualization

For single qubits, we can visualize the state space as a 3D sphere (Bloch sphere representation):

```javascript
console.log('Bloch Sphere Representation:');

console.log('\nSingle Qubit |ψ⟩ = α|0⟩ + β|1⟩ can be represented as:');
console.log('|ψ⟩ = cos(θ/2)|0⟩ + e^(iφ)sin(θ/2)|1⟩');
console.log('\nWhere θ and φ are spherical coordinates on the Bloch sphere.');

// Function to simulate Bloch vector calculations
function calculateBlochVector(stateName) {
    // Return theoretical Bloch vectors for common states
    if (stateName.includes('|0⟩')) return { x: 0, y: 0, z: 1 };
    if (stateName.includes('|1⟩')) return { x: 0, y: 0, z: -1 };
    if (stateName.includes('|+⟩')) return { x: 1, y: 0, z: 0 };
    if (stateName.includes('|-⟩')) return { x: -1, y: 0, z: 0 };
    if (stateName.includes('|+i⟩')) return { x: 0, y: 1, z: 0 };
    return { x: 0, y: 0, z: 1 };
}

// Function to visualize Bloch sphere state
function visualizeBlochState(circuit, stateName, description) {
    const state = circuit.execute();
    const bloch = calculateBlochVector(stateName);
    
    // Calculate spherical coordinates
    const r = Math.sqrt(bloch.x * bloch.x + bloch.y * bloch.y + bloch.z * bloch.z);
    const theta = Math.acos(bloch.z / r) * 180 / Math.PI;
    const phi = Math.atan2(bloch.y, bloch.x) * 180 / Math.PI;
    
    console.log(`\n${stateName}:`);
    console.log('━'.repeat(40));
    console.log('State representation:', state.toString());
    console.log(`Bloch Vector: (${bloch.x.toFixed(3)}, ${bloch.y.toFixed(3)}, ${bloch.z.toFixed(3)})`);
    console.log(`Spherical: θ=${theta.toFixed(0)}°, φ=${phi.toFixed(0)}° ${description}`);
    console.log('ASCII Bloch Sphere:');
    console.log('     ▲ Z');
    console.log('     ●  ← |0⟩');
    console.log('    /|\\');
    console.log('   / | \\');
    
    // Position marker based on state
    if (stateName.includes('|+⟩')) {
        console.log('  Y  ●→ X  ← |+⟩');
    } else if (stateName.includes('|-⟩')) {
        console.log(' Y ← ●  X  ← |-⟩');
    } else if (stateName.includes('|+i⟩')) {
        console.log('   ● | \\  ← |+i⟩');
    } else {
        console.log('  Y  |  X');
    }
    
    if (stateName.includes('|1⟩')) {
        console.log('     ●  ← |1⟩');
    } else {
        console.log('     ●  ← |1⟩');
    }
    console.log('━'.repeat(40));
}

// Visualize key Bloch sphere states
const states = [
    { name: 'State |0⟩', desc: '(North pole)', prep: [] },
    { name: 'State |1⟩', desc: '(South pole)', prep: ['x'] },
    { name: 'State |+⟩ = (|0⟩ + |1⟩)/√2', desc: '(Positive X-axis)', prep: ['h'] },
    { name: 'State |-⟩ = (|0⟩ - |1⟩)/√2', desc: '(Negative X-axis)', prep: ['x', 'h'] },
    { name: 'State |+i⟩ = (|0⟩ + i|1⟩)/√2', desc: '(Positive Y-axis)', prep: ['h', 's', 's', 's'] }
];

states.forEach(stateInfo => {
    const circuit = new Circuit(1); // Create circuit with 1 qubit
    
    // Apply preparation gates
    stateInfo.prep.forEach(gate => {
        if (gate === 'h') circuit.h(0); // lowercase h
        else if (gate === 'x') circuit.x(0); // lowercase x
        else if (gate === 's') circuit.s(0); // lowercase s
    });
    
    visualizeBlochState(circuit, stateInfo.name, stateInfo.desc);
});

console.log('\nBloch Sphere Properties:');
console.log('• Every single qubit state maps to a point on the sphere');
console.log('• Opposite points represent orthogonal states');
console.log('• Rotations correspond to unitary operations');
console.log('• Pure states are on the surface, mixed states inside');
console.log('• Cannot represent multi-qubit entanglement!');
```

## State Comparison and Fidelity

Let's create tools to compare quantum states and calculate fidelities:

```javascript
console.log('State Comparison and Fidelity:');

// Function to calculate theoretical fidelity between states
function calculateFidelity(stateName1, stateName2) {
    // Theoretical fidelity values for common state pairs
    if (stateName1 === stateName2) return 1.0;
    if ((stateName1.includes('|0⟩') && stateName2.includes('|1⟩')) ||
        (stateName1.includes('|1⟩') && stateName2.includes('|0⟩'))) return 0.0;
    if ((stateName1.includes('|0⟩') && stateName2.includes('|+⟩')) ||
        (stateName1.includes('|+⟩') && stateName2.includes('|0⟩'))) return 0.5;
    if ((stateName1.includes('|+⟩') && stateName2.includes('|-⟩')) ||
        (stateName1.includes('|-⟩') && stateName2.includes('|+⟩'))) return 0.0;
    return 0.25; // Default for Bell vs product states
}

// Function to compare two states
function compareStates(circuit1, circuit2, name1, name2) {
    const state1 = circuit1.execute();
    const state2 = circuit2.execute();
    const fidelity = calculateFidelity(name1, name2);
    
    console.log(`\nState A: ${name1}`);
    console.log('State A representation:', state1.toString());
    console.log(`State B: ${name2}`);
    console.log('State B representation:', state2.toString());
    console.log(`Fidelity F(A,B) = |⟨A|B⟩|² = ${fidelity.toFixed(3)} (${(fidelity * 100).toFixed(1)}%)`);
    
    if (fidelity > 0.99) {
        console.log('States are identical ✓');
    } else if (fidelity < 0.01) {
        console.log('States are orthogonal (completely different) ⊥');
    } else if (fidelity < 0.3) {
        console.log('States have low overlap');
    } else if (fidelity < 0.7) {
        console.log('States have moderate overlap');
    } else {
        console.log('States have high overlap');
    }
}

console.log('\nComparing Quantum States:');

// Create various states for comparison
const stateConfigs = [
    { name: '|0⟩', prep: [] },
    { name: '|0⟩ ', prep: [] },
    { name: '|1⟩', prep: ['x'] },
    { name: '|+⟩ = (|0⟩ + |1⟩)/√2', prep: ['h'] },
    { name: '|-⟩ = (|0⟩ - |1⟩)/√2', prep: ['x', 'h'] }
];

// Single qubit comparisons
const comparisons = [
    [0, 1], [0, 2], [0, 3], [3, 4]
];

comparisons.forEach(([i, j]) => {
    const circuit1 = new Circuit(1); // Create circuit with 1 qubit
    stateConfigs[i].prep.forEach(gate => {
        if (gate === 'h') circuit1.h(0); // lowercase h
        else if (gate === 'x') circuit1.x(0); // lowercase x
    });
    
    const circuit2 = new Circuit(1); // Create circuit with 1 qubit
    stateConfigs[j].prep.forEach(gate => {
        if (gate === 'h') circuit2.h(0); // lowercase h
        else if (gate === 'x') circuit2.x(0); // lowercase x
    });
    
    compareStates(circuit1, circuit2, stateConfigs[i].name, stateConfigs[j].name);
});

// Multi-qubit state comparisons
const bell1 = new Circuit(2); // Create circuit with 2 qubits
bell1.h(0); // lowercase h
bell1.cnot(0, 1); // lowercase cnot

const bell2 = new Circuit(2); // Create circuit with 2 qubits
bell2.h(0); // lowercase h
bell2.cnot(0, 1); // lowercase cnot

const product = new Circuit(2); // Create circuit with 2 qubits
product.h(0); // lowercase h
product.h(1); // lowercase h

compareStates(bell1, bell2, '|Φ+⟩ Bell state', '|Φ+⟩ Bell state');
compareStates(bell1, product, '|Φ+⟩ Bell state', '|++⟩ Product state');

console.log('\nFidelity Interpretation:');
console.log('• F = 1.0: Identical states');
console.log('• F = 0.0: Orthogonal states  ');
console.log('• 0 < F < 1: Partial overlap');
console.log('• F ≥ 0.99: High-fidelity quantum operations');
console.log('• F measures quantum state similarity');

console.log('\nDistance Metrics:');
console.log('• Trace Distance: D = (1 - F)/2 for pure states');
console.log('• Bures Distance: √(2(1 - √F))');
console.log('• Fidelity is most common in quantum computing');
```

## Summary

In this notebook, we explored comprehensive quantum state visualization techniques:

### Key Visualization Methods:
1. **Amplitude Display**: Complex numbers with magnitude and phase
2. **Probability Bars**: Visual representation of measurement probabilities 
3. **Multi-Qubit States**: High-dimensional state space visualization
4. **State Evolution**: Step-by-step transformation tracking
5. **Bloch Sphere**: 3D geometric representation for single qubits
6. **Fidelity Analysis**: Quantitative state comparison

### Important Concepts:
- **Amplitudes vs Probabilities**: $P(i) = |\alpha_i|^2$
- **Phase Information**: Complex phases affect interference
- **Entanglement**: Non-separable multi-qubit states
- **Bloch Vector**: $(x,y,z)$ coordinates for single qubits
- **Fidelity**: $F = |\langle\psi_1|\psi_2\rangle|^2$ measures state similarity

### Visualization Benefits:
1. **Intuitive Understanding**: Visual representation aids comprehension
2. **Error Detection**: Spot unexpected states or behaviors
3. **Algorithm Development**: Track state evolution through circuits
4. **Educational Value**: Teach quantum mechanics concepts
5. **Debugging Tools**: Identify issues in quantum programs

### Practical Applications:
- **Quantum Algorithm Design**: Visualize algorithm steps
- **State Preparation**: Verify target states are achieved
- **Error Analysis**: Measure deviations from ideal states
- **Educational Tools**: Demonstrate quantum phenomena
- **Research**: Explore new quantum states and properties

### Technical Details:
- **ASCII Graphics**: Text-based visualization for all platforms
- **Probability Bars**: Scaled visual representation
- **Phase Representation**: Polar form with magnitude and angle
- **Multi-dimensional**: Handle up to $2^n$ basis states
- **Real-time Updates**: Track state changes dynamically

State visualization is essential for understanding quantum computing, providing insights into the behavior of quantum systems that would be impossible to grasp from numerical data alone!
